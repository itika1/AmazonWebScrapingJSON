```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
import json
```

```python
html =('https://www.amazon.in/OnePlus-Mirror-Black-128GB-Storage/product-reviews/B07DJHV6VZ/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews')
```

```python
amazon_html = requests.get(html)
amazon_html
```

```python
amazon_html.content
```

```python
soup = bs(amazon_html.content,'html.parser')
```

```python
print(soup.prettify())
```

```python
titles = soup.find_all('div',class_='a-profile-content')
titles
```

```python
titles1 = soup.find_all('span',class_='a-profile-name')
titles1
```

```python
len(titles)
```

```python
customer_names =[]
for i in range(0, len(titles)):
    customer_names.append(titles[i].get_text())
customer_names
```

```python
final_customer_names= list(set(customer_names))
print(final_customer_names)
```

```python
name=soup.find_all('a',class_='review-title-content')
name
```

```python
reviews =[]
for i in range(0, len(name)):
    reviews.append(name[i].get_text())
reviews
```

```python
reviews[:]= [name.lstrip('\n') for name in reviews]
reviews[:]= [name.rstrip('\n') for name in reviews]
reviews
```

```python
len(reviews)
```
